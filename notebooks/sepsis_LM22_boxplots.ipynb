```
lm22=pd.read_table('/content/drive/My Drive/Mangul_Lab/Sepsis/summary_data/CombinedLM22.tsv.txt',sep='\t')
lm22.head()
```

```
def my_star(p_value):
    if p_value <= 0.001:
        return '***'
    elif p_value <= 0.01:
        return '**'
    elif p_value <= 0.05:
        return '*'
    else:
        return "None"
```

```
plt.figure(figsize=(8,8))
sns.set_style('white')
sns.set_context('talk')
age = sns.boxplot(x="survivor", y="age", data=lm22,order=[0,1])
x1, x2 = 0, 1
y, h, col = lm22['age'].max() + 4, 2, 'k'
plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
p_value = stats.mannwhitneyu(lm22[(lm22.survivor==0)].age.tolist(),lm22[(lm22.survivor==1)].age.tolist())
plt.text((x1+x2)*.5, y+h, my_star(p_value[1]), ha='center', va='bottom', color=col, fontsize=20)
#plt.yscale('log')
age.set_xticklabels(['Survived','Died'])
plt.xlabel("")
plt.ylabel("Age", fontsize=20)
#plt.savefig('../figures/age.png', bbox_inches='tight', pad_inches=.5)
plt.savefig("/content/drive/My Drive/Mangul_Lab/Sepsis/Figures/lm22_age_boxplot.png")
```
